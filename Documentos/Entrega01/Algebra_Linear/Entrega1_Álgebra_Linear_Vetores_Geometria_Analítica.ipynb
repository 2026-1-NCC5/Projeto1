# Criação base dados matricial normal

```python
import numpy as np
from PIL import Image
import pandas as pd
from google.colab import files
import io

print("Por favor, selecione a imagem dos donativos (Lideranças Empáticas) no seu computador...")

# 1. Abrir a janela para você escolher o arquivo
uploaded = files.upload()

if uploaded:
    caminho_da_imagem = next(iter(uploaded))
    image_bytes = uploaded[caminho_da_imagem]

    # 2. Carregar e processar a imagem (Tons de Cinza) usando BytesIO
    img = Image.open(io.BytesIO(image_bytes)).convert('L')

    # 3. Forçar o redimensionamento para 200x200
    # Isso garante que teremos exatamente 40.000 pontos de dados
    img_resul = img.resize((200, 200))

    # 4. Transformar em Matriz Numérica (Álgebra Linear na prática)
    matriz_pixels = np.array(img_resul)

    # 5. Salvar para importar no Google Sheets
    nome_saida = 'matriz_200x200_liderancas.csv'
    df = pd.DataFrame(matriz_pixels)
    df.to_csv(nome_saida, index=False, header=False)

    print(f"\n--- SUCESSO! ---")
    print(f"Arquivo '{nome_saida}' gerado com 40.000 pixels a partir de '{caminho_da_imagem}'.")
    print(f"Agora, importe este arquivo no Google Sheets.")
else:
    print("Nenhum arquivo foi selecionado.")
```

# Criação base dados matrical invertida

```python
import numpy as np
import pandas as pd
from google.colab import files
import io

print("Por favor, selecione o arquivo CSV a ser invertido...")

# 1. Abrir a janela para você escolher o arquivo CSV
uploaded = files.upload()

if uploaded:
    arquivo_original_nome = next(iter(uploaded))
    file_bytes = uploaded[arquivo_original_nome]

    try:
        # Lendo os dados como uma matriz numérica a partir dos bytes
        matriz = pd.read_csv(io.BytesIO(file_bytes), header=None).values

        # 2. A OPERAÇÃO MATRICIAL: Inverter as linhas (Flip Vertical)
        # Isso faz com que a primeira linha vire a última, a segunda vire a penúltima, etc.
        matriz_invertida = np.flipud(matriz)

        # 3. Salvar o novo arquivo CSV
        arquivo_saida = f'matriz_invertida_de_{arquivo_original_nome}'
        df_invertido = pd.DataFrame(matriz_invertida)
        df_invertido.to_csv(arquivo_saida, index=False, header=False)

        print(f"✅ Sucesso! O arquivo '{arquivo_saida}' foi gerado a partir de '{arquivo_original_nome}'.")
        print("Agora, baixe este arquivo na pastinha ao lado e importe no Google Sheets.")

    except Exception as e:
        print(f"❌ Erro ao processar o arquivo: {e}")
else:
    print("❌ Nenhum arquivo foi selecionado. Operação cancelada.")
```
